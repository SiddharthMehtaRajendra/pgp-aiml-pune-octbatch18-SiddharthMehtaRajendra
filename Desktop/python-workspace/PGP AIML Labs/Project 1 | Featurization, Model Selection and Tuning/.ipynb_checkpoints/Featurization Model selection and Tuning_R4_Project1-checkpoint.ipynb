Data is now fit for further processing

```python
# We will consider all columns for outlier analysis except id, price, waterfront and date as id and date are not valid for analysis 
# and price is our target variable. 
capDf = pd.DataFrame()
toCapDf = df.loc[:,:]
ineligible_cols = ['price', 'id', 'date', 'waterfront', 'view', 'yr_renovated']

for col in df.columns:
    if(col in ineligible_cols):
        capDf[col] = df[col]
        continue
    if(col in toCapDf.columns):
        percentiles = toCapDf[col].quantile([0.25,0.75]).values
        Q1 = percentiles[0]
        Q3 = percentiles[1]
        IQR = Q3 - Q1
        minCap = Q1 - (IQR) * 1.5
        maxCap = Q3 + (IQR) * 1.5
        capDf[col] = toCapDf[col][(toCapDf[col] >= minCap) & (toCapDf[col] <= maxCap)]
    else:
        capDf[col] = df[col]
    
capDf.dropna(inplace=True);
```

```python
#Printing the outliers which were removed (In all, 4902 eligible outlier rows were removed)
outlierDf = pd.concat([df, capDf])
outlierDf = outlierDf.reset_index(drop=True)
df_gpby = outlierDf.groupby(list(outlierDf.columns))
idx = [x[0] for x in df_gpby.groups.values() if len(x) == 1]
outlierDf.reindex(idx)
```

```python
#Outlier Capped Dataset
capDf.nunique()
```

```python
#Let us now drop the id column, which is irrelevant for analysis
capDf.drop(columns = 'id', axis = 1, inplace = True)
```

```python
capDf.info()
```

```python
#Since we require to correlate with mainly the status column (our target variable), we shall choose the same.
capDf.corr()['price']
```

```python
#Let us now drop variables from feature set having correlation in range -0.1 to 0.1 with target variable, if any.
X = capDf.iloc[:, capDf.columns != 'price']
Y = capDf[['price']]
corr_list = pd.Series(capDf.drop('price', axis=1).apply(lambda x: x.corr(capDf['price'])))
print("Correlation between Independent Variables and Target Variable (price): ")
print(corr_list)
for index, column in enumerate(X):
    if((corr_list[index] > -0.1) & (corr_list[index] < 0.1)):
        print("Variable Dropped: ", column)
        X.drop(columns=column, axis = 1, inplace=True)
```

```python
from sklearn import model_selection
from sklearn import preprocessing
```

Prior to performing linear regression, let us ensure the following assumptions or try to meet them:
1. Linearity
2. No Endogeneity
3. Normality and Homoscedasticity
4. No Autocorrelation
5. No Multicollinearity

```python
# Check for Linearity, plot each feature of X against Y, and ensure straight line fit
collated_data = pd.concat([X, Y], axis = 1)
for col in collated_data.columns:
    if(col != 'price'):
        sns.lmplot(x = col, y = 'price', data = collated_data)
```

Based off the Scatter plots above, the following features must be dropped as they violate the linearity principle:

1. Bedrooms
2. Bathrooms
3. Floors
4. Waterfront
5. View
6. Grade
7. yr_renovated

```python
linearity_compliant_data = X.drop(columns = ['bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'grade'
                                             , 'yr_renovated'], axis = 1)
#Adding 1 to each data point for better accuracy
for col in linearity_compliant_data.columns:
    linearity_compliant_data[col] = linearity_compliant_data[col] + 1
```

```python
linearity_compliant_data.head()
```

Check for endogeneity:

This is more concerned with intuition than with data wrangling. Now above, from after checking for linearity, we
observe that some variables dropped may affect price, for example the view and waterfront and also the Grade.

By intuition, I would like to include waterfront and view as a combined variable and the grade as another independent variable.

```python
endogeneity_complaint_data = pd.concat([linearity_compliant_data, X[["grade"]]], axis = 1)
waterfront_view = X[["waterfront", "view"]]
# Adding 1 to each variable for better results
endogeneity_complaint_data["WaterF_View"] = (waterfront_view.waterfront + 1) * (waterfront_view.view + 1)
endogeneity_complaint_data.head()
```

Check for Normality, Autocorrelation and multicollinearity:

1. Normalize the data
2. Autocorrelation (Not required here)
3. Multicollinearity (Shall skip as not seen as necessary)

```python
#Let us now finally standardize the data
test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = model_selection.train_test_split(endogeneity_complaint_data, Y, test_size=test_size, random_state=seed, shuffle = True)
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
print(scaler.mean_)
print(scaler.scale_)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
#Now for Linear Regression on Processed Dataset
from sklearn.linear_model import LinearRegression
model_e = LinearRegression()
model_e.fit(X_train_scaled, y_train)
```

```python
beta1 = model_e.coef_
print("Coef (b1): ", beta1)
coeffDf = pd.DataFrame(np.array(beta1).tolist(), columns = endogeneity_complaint_data.columns)
beta0 = model_e.intercept_
print("Intercept (b0): ", beta0)
```

```python
#Now for K-Fold Cross Validation
from sklearn.model_selection import KFold
num_folds = 10
seed = 77
k_fold = KFold(n_splits = num_folds,shuffle = True, random_state = seed)
```
