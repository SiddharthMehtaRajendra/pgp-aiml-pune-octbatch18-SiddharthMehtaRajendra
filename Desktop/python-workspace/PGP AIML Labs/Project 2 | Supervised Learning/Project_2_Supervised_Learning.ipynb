```python
#Replacing NaN values with 0
df.fillna(value = 0, axis = 1, inplace = True)
```

```python
#Capping Outliers in the given dataset. We shall only consider features which shall qualify for outlier analysis i.e.
#features with continuous values and not features that are categorical, order or count variables as they are ineligible for outlier
#analysis. Features which qualify for outlier analysis are Age, Experience, Income, Family Members, CCAvg and Mortgage
#in 'toCapDf' below.
capDf = pd.DataFrame()
toCapDf = df.iloc[:, [1, 2, 3, 5, 6, 8]]
for col in df.columns:
    if(df[col].dtype == object):
        capDf[col] = df[col]
        continue
    if(col in toCapDf.columns):
        percentiles = toCapDf[col].quantile([0.25,0.75]).values
        Q1 = percentiles[0]
        Q3 = percentiles[1]
        IQR = Q3 - Q1
        minCap = Q1 - (IQR) * 1.5
        maxCap = Q3 + (IQR) * 1.5
        capDf[col] = toCapDf[col][(toCapDf[col] >= minCap) & (toCapDf[col] <= maxCap)]
    else:
        capDf[col] = df[col]
    
capDf.dropna(inplace=True);
```

```python
#Printing the outliers which were removed (In all, 602 eligible outlier rows were removed)
outlierDf = pd.concat([df, capDf])
outlierDf = outlierDf.reset_index(drop=True)
df_gpby = outlierDf.groupby(list(outlierDf.columns))
idx = [x[0] for x in df_gpby.groups.values() if len(x) == 1]
outlierDf.reindex(idx)
```

Better than understanding the correlation values above and then applying limited knowledge to select independent variables, we shall use the automatic feature selection concept in Machine Learning to automate the task of selecting important features to analyse. The rationale behind this approach is to automate and also to improve the accuracy to selecting important features which can be effective candidates for our models to train. Note that we will be automating model selection on the independent variables and our dependent variable ('Personal Loan') shall remain separate.

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X = capDf.iloc[:, capDf.columns != 'Personal Loan']
Y = capDf['Personal Loan']

test_size = 0.20 # taking 70:30 training and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size)
```

We shall apply the Min Max Scaler on our data for allowing better analysis and for bringing the data to better homogenity for our models to train.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
select = SelectFromModel(
    RandomForestClassifier(n_estimators = 100, random_state = 42), threshold = 'median')
select.fit(X_train_scaled, y_train)
X_train_l1 = select.transform(X_train_scaled)
X_test_l1 = select.transform(X_test_scaled)
print("X_train.shape: {}".format(X_train_scaled.shape))
print("X_train_l1.shape: {}".format(X_train_l1.shape))
print("X_test_l1.shape: {}".format(X_test_l1.shape))
```

As we see above, of the 13 independent variables, 7 have been selected by the Model Based Feature Selector as effective candidates for our models to train with.

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train_l1,y_train)
```

```python
from sklearn.metrics import classification_report
predictions = logmodel.predict(X_test_l1)
print(classification_report(y_test,predictions))
```

```python
logmodel.score(X_test_l1, y_test)
```

Here, we see that our model has scored 95% which is a healthy model score.

```python
# calculate accuracy measures and confusion matrix
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, auc
```

```python
metrics.confusion_matrix(y_test, predictions)
```

```python
pd.DataFrame(data=[accuracy_score(y_test, predictions), 
                   recall_score(y_test, predictions),
                   precision_score(y_test, predictions)], 
             index=["accuracy", "recall", "precision"])
```

Here, we see that although the precision and accuracy of our model are high, the recall is low.

```python
from sklearn.neighbors import KNeighborsClassifier

MSE = {};
MSE_k = 0;
min_k = 1;
K = range(1,19)
for k in K:
    NNH = KNeighborsClassifier(n_neighbors= k)
    NNH.fit(X_train_l1, y_train.values.ravel())
    predicted_labels = NNH.predict(X_test_l1)
    MSE_k = 1 - accuracy_score(y_test, predicted_labels)
    MSE["{0}".format(k)]="{0}".format(MSE_k)
    
print(MSE)
print("Optimal Number of Neighbours and Corresponding MSE: {0}".format(min(MSE.items(), key=lambda x: x[1])))
```

Based on the above iterative processing to determine the optimum number of neighbours or 'k', we realize that the optimal number of neighbours which has the least corresponding Misclassification Error, is 1. Let us now check the accuracy for this model with k as 1.

```python
# Call Nearest Neighbour algorithm
NNH = KNeighborsClassifier(n_neighbors= 1)
NNH.fit(X_train_l1, y_train.values.ravel())
predicted_labels = NNH.predict(X_test_l1)
pd.DataFrame(data=[accuracy_score(y_test, predicted_labels), 
                   recall_score(y_test, predicted_labels),
                   precision_score(y_test, predicted_labels)], 
             index=["accuracy", "recall", "precision"])
```

We see above that not only the recall has improved significantly from around 46% to about 70% after utilizing the K Nearest Neighours algorithm, but also the accuracy has improved from about 95.7% to over 97%, and also the precision has increased from 90.6% to 95.6%.

```python

metrics.confusion_matrix(y_test, predicted_labels)
```

Above, we see the confusion matrix after applying the KNN model, where the increase in recall rate is starkly evident, as compared to applying the Logistic Regression Model.
